```python
import pandas as pd
import matplotlib.pyplot as plt

budget_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Budget': [50000, 120000, 80000, 60000, 90000]
}
df_budget = pd.DataFrame(budget_data)

expenses_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Actual_Expenses': [52000, 110000, 85000, 75000, 88000]
}
df_expenses = pd.DataFrame(expenses_data)

df_merged = pd.merge(df_budget, df_expenses, on='Department')
df_merged['Variance'] = df_merged['Actual_Expenses'] - df_merged['Budget']
df_merged['Exceeding_Budget'] = df_merged['Variance'] > 0
df_sorted = df_merged.sort_values(by='Variance', ascending=False)

print("\nMerged & Analyzed DataFrame:")
print(df_sorted)

plt.figure(figsize=(8, 5))
bar_width = 0.35
x = range(len(df_sorted))
plt.bar(x, df_sorted['Budget'], width=bar_width, label='Budget', alpha=0.7)
plt.bar([i + bar_width for i in x], df_sorted['Actual_Expenses'], width=bar_width, label='Actual Expenses', alpha=0.7)
plt.xticks([i + bar_width/2 for i in x], df_sorted['Department'])
plt.ylabel('Amount ($)')
plt.title('Department Budget vs Actual Expenses')
plt.legend()
plt.tight_layout()
plt.show()
```
