```python
import numpy as np

print("=== 2D Broadcasting Example ===")
A = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([10, 20, 30])

print("Array A:\n", A)
print("Shape of A:", A.shape)
print("Array b:", b)
print("Shape of b:", b.shape)

C = A + b
print("A + b =\n", C)
print("Result shape:", C.shape)
print()

print("=== Broadcasting Rules Explanation ===")
print("Shapes compared from right to left:")
print(" A.shape = (2, 3)")
print(" b.shape = (3,)")
print(" -> b is treated as (1, 3)")
print(" -> Dimensions align, so broadcasting works.")
print()

print("=== 3D Broadcasting with Scalar ===")
X = np.ones((2, 3, 4))
scalar = 5

print("Array X shape:", X.shape)
print("Scalar:", scalar)

Y = X + scalar
print("X + scalar =\n", Y)
print("Result shape:", Y.shape)
print("Explanation: Scalar is treated as shape (1, 1, 1).")
print()

print("=== Using np.newaxis ===")
d = np.array([1, 2, 3])

print("Original d shape:", d.shape)

d_col = d[:, np.newaxis]
print("d_col shape:", d_col.shape)
print("d_col:\n", d_col)

M = np.array([[10, 20, 30],
                            [40, 50, 60],
                            [70, 80, 90]])

print("M shape:", M.shape)

result = M + d_col
print("M + d_col =\n", result)
print("Result shape:", result.shape)
print("Explanation: d_col (3,1) is broadcast across columns.")
print()

print("=== Invalid Broadcasting Case ===")
P = np.ones((2, 3))
Q = np.ones((3, 2))

print("P shape:", P.shape)
print("Q shape:", Q.shape)

try:
    Z = P + Q
except ValueError as e:
    print("Error:", e)
    print("Explanation: Shapes (2,3) and (3,2) are incompatible.")
```
